```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('data_eprc.csv')
```

#### Histogram

```python
sns.histplot(data.price)
```

#### Customize the bins

```python
sns.histplot(data.price, bins=np.arange(0,20,1))
```

#### Styles (You can also use the styles in other plots such as lineplot, scatterplot...)

```python
plt.style.available
```

```python
plt.style.use('ggplot')        # switch system's style to ggplot
sns.histplot(data.price, bins=np.arange(0,20,1))
plt.style.use('default')       # switch system's style to default
```

```python
plt.style.use(['ggplot','dark_background'])        # use a combination of multiple styles
sns.histplot(data.price, bins=np.arange(0,20,1))
plt.style.use('default')       # switch system's style to default
```

```python
plt.xkcd()           # twist the lines if you like
sns.histplot(data.price, bins=np.arange(0,20,1))
plt.rcdefaults()     # swtich back to default settings
```

#### Lines and text 

```python
#plt.figure(dpi=100)
sns.histplot(data.price, bins=np.arange(0,20,1))
# draw lines
plt.vlines(data.price.mean(), ymin=0, ymax=30000, color='green')
plt.vlines(data.price.median(), ymin=0, ymax=30000, color='red')     # plt.hlines() is for horizontal lines
# draw text
plt.text(x = 5, y = 20000, s = 'mean > median', color='black', fontsize=15)
plt.text(x = 5, y = 15000, s = 'skewness > 0 (right-skewed)', color='black', fontsize=15)
```

#### Common descriptive statistics 

```python
data.price.max()
data.price.min()
data.price.mean()
data.price.std()
data.price.skew()
data.price.kurtosis()
data.price.quantile(0.5)
data.price.median()
```


### <font color='red'>Tip 3: How to explore more about this variable?</font>
>  It is usually helpful to zoom in and zoom out by changing the bins.


```python
sns.histplot(data.price, bins=np.arange(1,3,0.02))
```

```python
sns.histplot(data.price, bins=np.arange(7,9,0.02))
```

```python
plt.figure(figsize=(10,5),dpi=100)

plt.subplot(1,2,1)
sns.histplot(data.price, bins=np.arange(1,3,0.02))

plt.subplot(1,2,2)
sns.histplot(data.price, bins=np.arange(7,9,0.02))
```
